```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
real_estate_property_list = pd.read_csv('histogram_data.csv')
df = real_estate_property_list.copy()
df
```

```python
sns.set_style("white")
plt.figure(figsize=(8,6))
plt.hist(df['Price'], bins=8, color = '#108A99')
plt.title("Distribution of Real Estate Prices", fontsize = 14, weight = 'bold')
plt.xlabel("Price in (000' $)")
plt.ylabel('Number of Properties')
sns.despine()
plt.show()
```
