```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('stacked_area_chart_data.csv')
df = raw_data.copy()
df
```

```python
colors = ['#011638', '#7e2987', '#ef2026']
labels = ['Diesel', 'Petrol', 'Gas']
plt.figure(figsize=(12,6))
plt.stackplot(df['Year'], df['Diesel'], df['Petrol'], df['Gas'], colors = colors, edgecolor = 'none')
plt.xticks(df['Year'], rotation = 45)
plt.legend(labels = labels, loc = 'upper left')
plt.ylabel('Number of Cars', fontsize = 13)
plt.title("Popularity of Engine fuel Type(1982 - 2016)", fontsize = 14, weight = 'bold')
# despine() from seaborn allows us to remove the top and right border of the chart
sns.despine()
plt.show()
```
