```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('scatter_plot_ii.csv')
df = raw_data.copy()
df
```

```python
#plt.figure(figsize=(10,8))
sns.set(rc={'figure.figsize':(9,6)})
sns.regplot(x = 'Budget', y = 'Sales', data = df, scatter_kws = {'color':'k'}, line_kws = {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's $)", fontsize=14)
plt.ylabel("Sales in (000's Units)", fontsize=14)
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, weight='bold')
plt.show()
```

```python
# 'height' sets both the height and width of the plot
sns.lmplot(x='Budget', y='Sales', data=df, height = 10, scatter_kws = {'color':'k'}, line_kws = {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's $)", fontsize=14)
plt.ylabel("Sales in (000's Units)", fontsize=14)
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, weight='bold')
plt.show()
```
